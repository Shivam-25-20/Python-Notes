# Characteristics of Lists

Python lists have several important characteristics that make them one of the most flexible and widely used data structures.

The main characteristics of a list are:

1. Ordered
2. Mutable
3. Allows Duplicates
4. Heterogeneous
5. Dynamic Size

---

## 1. Ordered

Lists are **ordered**, which means the elements are stored in a specific sequence.

Each element has a fixed position called an **index**.

### Example

````python
fruits = ["Apple", "Banana", "Mango"]

print(fruits)
````

**Output**

```text
['Apple', 'Banana', 'Mango']
```

**Explanation**

The elements remain in the same order in which they were added.

---

## 2. Mutable

Lists are **mutable**, which means their elements can be changed after the list is created.

You can:

- Modify elements
- Add elements
- Remove elements

### Example

````python
fruits = ["Apple", "Banana", "Mango"]

fruits[1] = "Orange"

print(fruits)
````

**Output**

```text
['Apple', 'Orange', 'Mango']
```

**Explanation**

The second element `"Banana"` is replaced with `"Orange"`.

---

## 3. Allows Duplicates

Lists allow duplicate values.

The same element can appear multiple times.

### Example

````python
numbers = [10, 20, 20, 30, 30]

print(numbers)
````

**Output**

```text
[10, 20, 20, 30, 30]
```

**Explanation**

The values `20` and `30` appear more than once, and Python stores them without any problem.

---

## 4. Heterogeneous

A list can store elements of **different data types**.

For example, a single list can contain:

- Integers
- Strings
- Floating-point numbers
- Boolean values
- Even other lists

### Example

````python
data = ["Alice", 20, 95.5, True]

print(data)
````

**Output**

```text
['Alice', 20, 95.5, True]
```

**Explanation**

The list contains four different data types.

---

## 5. Dynamic Size

Lists are **dynamic**, meaning their size can increase or decrease during program execution.

You can add or remove elements whenever needed.

### Example

````python
numbers = [10, 20, 30]

numbers.append(40)

print(numbers)
````

**Output**

```text
[10, 20, 30, 40]
```

**Explanation**

The new element `40` is added to the end of the list, increasing its size.

---

## Summary Table

| Characteristic | Description |
|----------------|-------------|
| Ordered | Elements maintain their insertion order. |
| Mutable | Elements can be changed after creation. |
| Allows Duplicates | Duplicate values are permitted. |
| Heterogeneous | Different data types can be stored together. |
| Dynamic Size | Elements can be added or removed at any time. |

---

## Key Points

- Lists preserve the order of elements.
- Lists are mutable and can be modified.
- Duplicate elements are allowed.
- Lists can contain multiple data types.
- Lists automatically grow or shrink as needed.

# List Slicing

**List slicing** is used to access a portion (or subset) of a list.

Instead of accessing a single element, slicing allows you to retrieve **multiple elements** at once.

> **Note:** Slicing creates a **new list** without modifying the original list.

---

## Syntax

```python
list_name[start : stop : step]
```

### Syntax Explanation

- **start** → Starting index (inclusive).
- **stop** → Ending index (exclusive).
- **step** → Number of positions to move (optional).

If any value is omitted, Python uses a default value.

---

## Understanding the Indexes

Consider the following list:

```python
fruits = ["Apple", "Banana", "Mango", "Orange", "Grapes"]
```

| Positive Index | 0 | 1 | 2 | 3 | 4 |
|:--------------:|:-:|:-:|:-:|:-:|:-:|
| Value | Apple | Banana | Mango | Orange | Grapes |
| Negative Index | -5 | -4 | -3 | -2 | -1 |

---

## 1. Basic Slicing

### Example

````python
fruits = ["Apple", "Banana", "Mango", "Orange", "Grapes"]

print(fruits[1:4])
````

**Output**

```text
['Banana', 'Mango', 'Orange']
```

**Explanation**

- Starts from index `1`.
- Stops before index `4`.
- Index `4` is **not included**.

---

## 2. Omitting the Start Index

If the start index is omitted, slicing begins from the first element.

### Example

````python
fruits = ["Apple", "Banana", "Mango", "Orange", "Grapes"]

print(fruits[:3])
````

**Output**

```text
['Apple', 'Banana', 'Mango']
```

---

## 3. Omitting the Stop Index

If the stop index is omitted, slicing continues until the end of the list.

### Example

````python
fruits = ["Apple", "Banana", "Mango", "Orange", "Grapes"]

print(fruits[2:])
````

**Output**

```text
['Mango', 'Orange', 'Grapes']
```

---

## 4. Omitting Both Start and Stop

### Example

````python
fruits = ["Apple", "Banana", "Mango"]

print(fruits[:])
````

**Output**

```text
['Apple', 'Banana', 'Mango']
```

**Explanation**

This returns a copy of the entire list.

---

## 5. Using Step

The **step** value determines how many positions Python moves after selecting an element.

### Example

````python
numbers = [10, 20, 30, 40, 50, 60]

print(numbers[::2])
````

**Output**

```text
[10, 30, 50]
```

**Explanation**

Python selects every second element.

---

### Example

````python
numbers = [10, 20, 30, 40, 50, 60]

print(numbers[1::2])
````

**Output**

```text
[20, 40, 60]
```

---

## 6. Negative Slicing

Negative indexes can also be used while slicing.

### Example

````python
fruits = ["Apple", "Banana", "Mango", "Orange", "Grapes"]

print(fruits[-4:-1])
````

**Output**

```text
['Banana', 'Mango', 'Orange']
```

**Explanation**

- Starts from `-4`.
- Stops before `-1`.

---

## 7. Reversing a List

A negative step reverses the direction of slicing.

### Example

````python
numbers = [10, 20, 30, 40, 50]

print(numbers[::-1])
````

**Output**

```text
[50, 40, 30, 20, 10]
```

**Explanation**

`[::-1]` creates a reversed copy of the list.

---

## 8. Copying a List Using Slicing

### Example

````python
numbers = [10, 20, 30]

copy_numbers = numbers[:]

print(copy_numbers)
````

**Output**

```text
[10, 20, 30]
```

**Explanation**

Slicing with `[:]` creates a shallow copy of the list.

---

## Summary Table

| Expression | Description |
|------------|-------------|
| `list[:]` | Copy the entire list |
| `list[1:4]` | Elements from index 1 to 3 |
| `list[:3]` | First three elements |
| `list[2:]` | From index 2 to the end |
| `list[::2]` | Every second element |
| `list[::-1]` | Reverse the list |
| `list[-4:-1]` | Slice using negative indexes |

## Key Points

- Slicing returns a new list.
- The **start index is included**.

# Looping Through Lists

Looping through a list means accessing each element of the list one by one.

Python provides several ways to iterate over a list.

The most common methods are:

- `for` loop
- `for` loop with `range()`
- `while` loop
- `enumerate()`

> **Note:** The `for` loop is the most commonly used and recommended way to iterate through a list.

---

# 1. Using a `for` Loop

The `for` loop directly accesses each element of the list.

## Syntax

```python
for item in list_name:
    # Code
```

### Example

````python
fruits = ["Apple", "Banana", "Mango"]

for fruit in fruits:
    print(fruit)
````

**Output**

```text
Apple
Banana
Mango
```

**Explanation**

The loop visits each element one by one and prints it.

---

# 2. Using `range()`

The `range()` function iterates through the indexes of the list.

## Syntax

```python
for index in range(len(list_name)):
    # Code
```

### Example

````python
fruits = ["Apple", "Banana", "Mango"]

for i in range(len(fruits)):
    print(fruits[i])
````

**Output**

```text
Apple
Banana
Mango
```

**Explanation**

- `len(fruits)` returns the number of elements.
- `range()` generates the indexes.
- `fruits[i]` accesses each element using its index.

---

# 3. Using a `while` Loop

A `while` loop can also be used to iterate through a list.

### Example

````python
fruits = ["Apple", "Banana", "Mango"]

i = 0

while i < len(fruits):
    print(fruits[i])
    i += 1
````

**Output**

```text
Apple
Banana
Mango
```

**Explanation**

The loop continues until all elements have been accessed.

---

# 4. Using `enumerate()`

The `enumerate()` function returns both the **index** and the **value** of each element.

### Example

````python
fruits = ["Apple", "Banana", "Mango"]

for index, value in enumerate(fruits):
    print(index, value)
````

**Output**

```text
0 Apple
1 Banana
2 Mango
```

**Explanation**

- `index` stores the position of the element.
- `value` stores the actual element.

`enumerate()` is useful when you need both the index and the value.

---

# Summary Table

| Method | When to Use |
|---------|-------------|
| `for` loop | When only the elements are needed. |
| `range()` | When the indexes are needed. |
| `while` loop | When iteration depends on a condition. |
| `enumerate()` | When both the index and the element are needed. |

---

## Key Points

- The `for` loop is the simplest way to iterate through a list.
- Use `range(len(list))` when you need indexes.
- Use a `while` loop when iteration depends on a condition.
- Use `enumerate()` when you need both the index and the value.
- All these methods iterate through every element in the list.

# Copying Lists

Sometimes you need to create a copy of a list instead of working with the original list.

Python provides several ways to copy a list.

The most common methods are:

- Assignment (`=`)
- `copy()` method
- Slicing (`[:]`)

> **Note:** Assignment (`=`) does **not** create a new list. It creates another reference to the same list.

---

# 1. Using Assignment (`=`)

When you assign one list to another using `=`, both variables refer to the **same list**.

## Example

````python
list1 = [10, 20, 30]

list2 = list1

list2[0] = 100

print("List 1:", list1)
print("List 2:", list2)
````

**Output**

```text
List 1: [100, 20, 30]
List 2: [100, 20, 30]
```

**Explanation**

- `list1` and `list2` refer to the same list.
- Changing one list also changes the other.

---

# 2. Using the `copy()` Method

The `copy()` method creates a **shallow copy** of the list.

The copied list is a different object, so modifying it does not affect the original list.

## Syntax

```python
new_list = list_name.copy()
```

### Example

````python
list1 = [10, 20, 30]

list2 = list1.copy()

list2[0] = 100

print("List 1:", list1)
print("List 2:", list2)
````

**Output**

```text
List 1: [10, 20, 30]
List 2: [100, 20, 30]
```

**Explanation**

`copy()` creates a new list with the same elements.

---

# 3. Using List Slicing (`[:]`)

A list can also be copied using slicing.

## Syntax

```python
new_list = list_name[:]
```

### Example

````python
list1 = [10, 20, 30]

list2 = list1[:]

list2[1] = 200

print("List 1:", list1)
print("List 2:", list2)
````

**Output**

```text
List 1: [10, 20, 30]
List 2: [10, 200, 30]
```

**Explanation**

`[:]` creates a new list containing the same elements.

---

# Assignment vs `copy()` vs Slicing

| Method | Creates a New List | Changes Affect Original List |
|--------|--------------------|------------------------------|
| `=` | ❌ No | ✅ Yes |
| `copy()` | ✅ Yes | ❌ No |
| `[:]` | ✅ Yes | ❌ No |

---

## Which Method Should You Use?

| Method | Recommended Use |
|---------|-----------------|
| `=` | When both variables should refer to the same list. |
| `copy()` | Best way to create a copy of a list. |
| `[:]` | Quick way to create a copy using slicing. |

---

## Key Points

- `=` creates another reference to the same list.
- `copy()` creates a shallow copy of the list.
- `[:]` also creates a shallow copy.
- Modifying a copied list does not affect the original list.
- Use `copy()` or `[:]` when you need an independent copy of a list.

# Joining Lists

Joining lists means combining two or more lists into a single list.

Python provides several ways to join lists.

The most common methods are:

- Using the `+` operator
- Using the `extend()` method
- Using a `for` loop

> **Note:** Choose the method based on whether you want to create a new list or modify an existing list.

---

# 1. Using the `+` Operator

The `+` operator combines two or more lists and returns a **new list**.

## Syntax

```python
new_list = list1 + list2
```

### Example

````python
list1 = [1, 2, 3]
list2 = [4, 5, 6]

result = list1 + list2

print(result)
````

**Output**

```text
[1, 2, 3, 4, 5, 6]
```

**Explanation**

The `+` operator joins both lists and creates a new list. The original lists remain unchanged.

---

# 2. Using the `extend()` Method

The `extend()` method adds all elements of one list to another.

## Syntax

```python
list1.extend(list2)
```

### Example

````python
list1 = [1, 2, 3]
list2 = [4, 5, 6]

list1.extend(list2)

print(list1)
````

**Output**

```text
[1, 2, 3, 4, 5, 6]
```

**Explanation**

The elements of `list2` are added to `list1`.

Unlike the `+` operator, `extend()` modifies the original list.

---

# 3. Using a `for` Loop

You can also join lists by adding each element individually using a loop.

### Example

````python
list1 = [1, 2, 3]
list2 = [4, 5, 6]

for item in list2:
    list1.append(item)

print(list1)
````

**Output**

```text
[1, 2, 3, 4, 5, 6]
```

**Explanation**

The loop iterates through `list2` and adds each element to `list1` using `append()`.

---

# Comparison of Joining Methods

| Method | Creates a New List | Modifies Original List |
|---------|--------------------|------------------------|
| `+` | ✅ Yes | ❌ No |
| `extend()` | ❌ No | ✅ Yes |
| `for` loop + `append()` | ❌ No | ✅ Yes |

---

## Which Method Should You Use?

| Method | Best Use |
|---------|----------|
| `+` | When you want a new combined list. |
| `extend()` | When you want to add elements to an existing list. |
| `for` loop | When you need additional processing while joining lists. |

---
| Operation             | Result                                        |
| --------------------- | --------------------------------------------- |
| `list1 + list2`       | Creates a **new list**                        |
| `list1.append(list2)` | Adds **list2 as a single nested element**     |
| `list1.extend(list2)` | Adds **each element** of `list2` individually |
---

## Key Points

- The `+` operator joins lists and returns a new list.
- The `extend()` method adds all elements of one list to another.
- A `for` loop with `append()` can also be used to join lists.
- `extend()` and the `for` loop modify the original list.
- The `+` operator keeps the original lists unchanged.

# Nested Lists

A **nested list** is a list that contains one or more lists as its elements.

In other words, a list inside another list is called a **nested list**.

Nested lists are useful for storing data in a table-like structure, such as student records, marks, or matrices.

---

## Syntax

```python
list_name = [[item1, item2], [item3, item4]]
```

---

## Example 1: Creating a Nested List

````python
students = [
    ["Alice", 20],
    ["Bob", 21],
    ["Charlie", 22]
]

print(students)
````

**Output**

```text
[['Alice', 20], ['Bob', 21], ['Charlie', 22]]
```

**Explanation**

The main list contains three sublists.

Each sublist stores the name and age of a student.

---

## Structure of a Nested List

```python
students = [
    ["Alice", 20],
    ["Bob", 21],
    ["Charlie", 22]
]
```

| Main Index | Sublist |
|------------|---------|
| `0` | `["Alice", 20]` |
| `1` | `["Bob", 21]` |
| `2` | `["Charlie", 22]` |

---

## Accessing Nested List Elements

To access an element inside a nested list, use **multiple indexes**.

## Syntax

```python
list_name[outer_index][inner_index]
```

- `outer_index` → Selects the sublist.
- `inner_index` → Selects the element inside the sublist.

---

## Example 2: Accessing a Sublist

````python
students = [
    ["Alice", 20],
    ["Bob", 21],
    ["Charlie", 22]
]

print(students[1])
````

**Output**

```text
['Bob', 21]
```

---

## Example 3: Accessing an Individual Element

````python
students = [
    ["Alice", 20],
    ["Bob", 21],
    ["Charlie", 22]
]

print(students[1][0])
````

**Output**

```text
Bob
```

**Explanation**

- `students[1]` selects the second sublist.
- `[0]` selects the first element of that sublist.

---

## Example 4: Accessing Another Element

````python
students = [
    ["Alice", 20],
    ["Bob", 21],
    ["Charlie", 22]
]

print(students[2][1])
````

**Output**

```text
22
```

---

## Modifying Nested List Elements

Nested list elements can be modified using indexes.

### Example

````python
students = [
    ["Alice", 20],
    ["Bob", 21],
    ["Charlie", 22]
]

students[1][0] = "David"

print(students)
````

**Output**

```text
[['Alice', 20], ['David', 21], ['Charlie', 22]]
```

**Explanation**

The name `"Bob"` is replaced with `"David"`.

---

## Looping Through a Nested List

Nested loops are commonly used to access every element.

### Example

````python
students = [
    ["Alice", 20],
    ["Bob", 21],
    ["Charlie", 22]
]

for row in students:
    for item in row:
        print(item)
````

**Output**

```text
Alice
20
Bob
21
Charlie
22
```

**Explanation**

- The outer loop accesses each sublist.
- The inner loop accesses each element inside the sublist.

---

## Real-Life Example

````python
marks = [
    ["Alice", 85],
    ["Bob", 92],
    ["Charlie", 78]
]

for student in marks:
    print(student)
````

**Output**

```text
['Alice', 85]
['Bob', 92]
['Charlie', 78]
```

---

## Summary Table

|---------|---------|
| Create a nested list | `[[1, 2], [3, 4]]` |
| Access a sublist | `list[0]` |
| Access an element | `list[0][1]` |
| Modify an element | `list[1][0] = value` |
| Loop through nested lists | Nested `for` loops |

---

## Key Points

- A nested list is a list containing one or more lists.
- Access nested elements using multiple indexes.
- Nested lists are mutable, so their elements can be modified.
- Nested loops are commonly used to iterate through nested lists.
- Nested lists are useful for representing tables, matrices, and structured data.

# Introduction to Matrix

A **matrix** is a rectangular arrangement of data organized into **rows** and **columns**.

In Python, a matrix is commonly represented using a **nested list**, where each inner list represents a row of the matrix.

> **Note:** Python does not have a built-in matrix data type. Beginners usually represent matrices using nested lists. For advanced matrix operations, the **NumPy** library is commonly used.

---

## Example of a Matrix

````python
matrix = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

print(matrix)
````

**Output**

```text
[[1, 2, 3],
 [4, 5, 6],
 [7, 8, 9]]
```

---

## Matrix Representation

```text
1  2  3

7  8  9
```

- Number of rows = **3**
- Number of columns = **3**

---

## Accessing Matrix Elements

Since a matrix is a nested list, elements are accessed using two indexes.

### Syntax

```python
matrix[row][column]
```

### Example

````python
matrix = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

print(matrix[1][2])
````

**Output**

```text
6
```

**Explanation**

- `matrix[1]` selects the second row: `[4, 5, 6]`
- `matrix[1][2]` selects the third element of that row, which is `6`.

- Game Development

## Key Points

- A matrix is a collection of rows and columns.

- For advanced matrix operations, the **NumPy** library is commonly used.

# List Comprehension (Introduction)

A **list comprehension** is a short and concise way to create a new list in Python.

It allows you to generate a list using a single line of code instead of writing multiple lines with a loop.

> **Note:** List comprehensions make code shorter and more readable. However, they should only be used when they improve readability.

---

## Without List Comprehension

````python
numbers = []

for i in range(5):
    numbers.append(i)

print(numbers)
````

**Output**

```text
[0, 1, 2, 3, 4]
```

---

## With List Comprehension

````python
numbers = [i for i in range(5)]

print(numbers)
````

**Output**

```text
[0, 1, 2, 3, 4]
```

**Explanation**

Both programs produce the same result, but the list comprehension version is shorter.

---

# Syntax

```python
new_list = [expression for item in iterable]
```

### Syntax Explanation

- **expression** → The value to be added to the new list.
- **item** → Variable used to access each element.
- **iterable** → A sequence such as a list, tuple, string, or `range()`.

---

# Example 1: Create a List of Numbers

````python
numbers = [x for x in range(1, 6)]

print(numbers)
````

**Output**

```text
[1, 2, 3, 4, 5]
```

---

# Example 2: Square of Numbers

````python
squares = [x * x for x in range(1, 6)]

print(squares)
````

**Output**

```text
[1, 4, 9, 16, 25]
```

**Explanation**

Each number is squared before being added to the list.

---

# List Comprehension with `if`

A condition can be used to include only specific elements.

## Syntax

```python
new_list = [expression for item in iterable if condition]
```

### Example

````python
even_numbers = [x for x in range(1, 11) if x % 2 == 0]

print(even_numbers)
````

**Output**

```text
[2, 4, 6, 8, 10]
```

**Explanation**

Only even numbers are added to the list.

---

# List Comprehension with `if...else`

You can also use `if...else` inside a list comprehension.

## Syntax

```python
new_list = [expression1 if condition else expression2 for item in iterable]
```

### Example

````python
result = ["Even" if x % 2 == 0 else "Odd" for x in range(1, 6)]

print(result)
````

**Output**

```text
['Odd', 'Even', 'Odd', 'Even', 'Odd']
```

**Explanation**

For each number:

- If it is even, `"Even"` is added.
- Otherwise, `"Odd"` is added.

---

# Advantages of List Comprehension

- Shorter and cleaner code.
- Easy to create new lists.
- More readable for simple operations.
- Often faster than using a `for` loop.

---

# Limitations

- Can become difficult to read if too complex.
- Not suitable for large or complicated logic.
- Use a normal `for` loop when readability is more important.

---

# Summary Table

| Feature | Example |
|---------|---------|
| Basic | `[x for x in range(5)]` |
| Expression | `[x * x for x in range(5)]` |
| With `if` | `[x for x in range(10) if x % 2 == 0]` |
| With `if...else` | `["Even" if x % 2 == 0 else "Odd" for x in range(5)]` |

---

## Key Points

- List comprehension creates a new list in a single line.
- It is an alternative to using a `for` loop with `append()`.
- It can include expressions and conditions.
- Use list comprehensions for simple and readable code.
- Use a normal loop for complex logic.

# Common Errors

- Using an invalid index (`IndexError`).
- Confusing `append()` and `extend()`.
- Using `remove()` for an element that does not exist (`ValueError`).
- Forgetting that list indexing starts from `0`.
- Using `=` instead of `copy()` when copying a list.
- Mixing up `sort()` and `sorted()`.
- Incorrect slicing (`start:stop:step`).
- Modifying a list while iterating over it.

---

# Best Practices

- Use meaningful list variable names.
- Use `append()` to add one element and `extend()` to add multiple elements.
- Use `copy()` or `[:]` to create an independent copy of a list.
- Check that an index exists before accessing it.
- Use list comprehensions for simple and readable code.
- Keep list operations simple and easy to understand.
- Avoid modifying a list while looping through it.
- Choose the appropriate list method for the task.
