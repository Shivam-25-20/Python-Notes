# History of Python

Python is a high-level, interpreted programming language created by Guido van Rossum. Development of Python began in the late 1980s, and its first official version, Python 0.9.0, was released in 1991.

Guido van Rossum wanted to create a programming language that was simple, easy to read, and powerful. He named it Python after the British comedy show "Monty Python's Flying Circus", not after the snake.

Over the years, Python has become one of the most popular programming languages due to its simple syntax and wide range of applications.

| Object-Oriented | Yes | No |

# Comments in Python

Comments are notes added to a Python program to explain the code. They are ignored by the Python interpreter and are not executed.

---

## Types of Comments

### 1. Single-Line Comment

A single-line comment starts with the `#` symbol.

**Example:**

```python
# This is a single-line comment
print("Hello, World!")
```

---

### 2. Multi-Line Comment

Python does not have a separate syntax for multi-line comments. Triple quotes (`'''` or `"""`) are commonly used to write multi-line descriptions.

**Example:**

```python
"""
This is a
multi-line comment.
"""

print("Python")
```

---
You can paste this directly into a Markdown cell.

## 3. Inline Comment

An **inline comment** is written on the same line as a Python statement. It is used to explain a specific part of the code.

### Syntax

```python
statement  # Comment
```

### Example

```python
print("Hello, World!")  # Prints a greeting message
```
---
## Why Use Comments?

- Explain the purpose of the code.
- Make the program easier to understand.
- Help during debugging and maintenance.
- Improve code readability.

---

## Key Points

- Single-line comments begin with `#`.
- Multi-line comments can be written using triple quotes (`'''` or `"""`).
- Comments are ignored during program execution.

# Common Errors in Python

Beginners often make mistakes while writing Python programs. These mistakes are called **errors**. Understanding common errors helps in debugging programs.

---

## 1. Syntax Error

A **Syntax Error** occurs when Python code does not follow the correct syntax.

**Example:**

```python
print("Hello"
```

**Error:**

```text
SyntaxError: '(' was never closed
```

---

## 2. Name Error

A **Name Error** occurs when a variable or function is used before it is defined.

**Example:**

```python
print(message)
```

**Error:**

```text
NameError: name 'message' is not defined
```

---

## 3. Indentation Error

An **Indentation Error** occurs when the required indentation is missing or incorrect.

**Example:**

```python
if True:
print("Hello")
```

**Error:**

```text
IndentationError: expected an indented block
```

---

## 4. Type Error

A **Type Error** occurs when incompatible data types are used together.

**Example:**

```python
print("Age: " + 20)
```

**Error:**

```text
TypeError
```

---

## 5. Zero Division Error

A **ZeroDivisionError** occurs when a number is divided by zero.

**Example:**

```python
print(10 / 0)
```

**Error:**

```text
ZeroDivisionError: division by zero
```

---

## Summary Table

| Error | Cause |
|--------|-------|
| SyntaxError | Incorrect Python syntax |
| NameError | Undefined variable or function |
| IndentationError | Incorrect indentation |
| TypeError | Incompatible data types |
| ZeroDivisionError | Division by zero |

---

## Key Points

- Errors are common while learning Python.
- Read the error message carefully.
- Correct the mistake and run the program again.
